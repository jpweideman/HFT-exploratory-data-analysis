```python
#concatenate exec_data, trades_data and spreads_data 
data_concatenated = pd.concat([exec_data, trades_data, spreads_data])
```

```python
# make the setting so all columns are displayed
pd.set_option('display.max_columns', None)
```

```python
#remove unnamed:0 column
data_concatenated = data_concatenated.loc[:, ~data_concatenated.columns.str.contains('^Unnamed')]
```

```python
#export the concatenated data to a csv file
data_concatenated.to_csv('Data/concatenated_data.csv', index=False)
```

```python
# concatenate exec_data, trades_data and spreads_data so that the rows are mathed by ObjectId and Date
data_concatenated_grouped = data_concatenated.groupby(['ObjectId', 'Date'])
```

```python
#display the dataset data_concatenated_grouped
data_concatenated_grouped.first()
```

```python
# make a dataframe from the data_concatenated_grouped
data_concatenated_grouped = pd.DataFrame(data_concatenated_grouped.first())
```

```python
# Convert grouped object back to a DataFrame
grouped_df = data_concatenated_grouped.apply(lambda x: x)

# Reset index to convert groupby object to a regular DataFrame with multi-level index as columns
grouped_df = grouped_df.reset_index(drop=True)

```
