```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import all datasets in Data folder
exec_data = pd.read_csv('Data/1. exec_data_01-01-2024.csv')
trades_data = pd.read_csv('Data/2. trades_data_01-01-2024.csv')
spreads_data = pd.read_csv('Data/3. spreads_data_01-01-2024.csv')
listingId_data = pd.read_csv('Data/listingId_ref_data_01-01-2024.csv')
market_data = pd.read_csv('Data/markets_ref_data_01-01-2024.csv')
metrics_data = pd.read_csv('Data/metrics_ref_data_01-01-2024.csv')
```

```python
# Merging datasets on common columns 'ListingId' and 'Date'
merged_df = exec_data.merge(trades_data, on=['ObjectId', 'Date'], how='outer').merge(spreads_data, on=['ObjectId', 'Date'], how='outer')
```

```python
#save the merged data to a csv file
merged_df.to_csv('Data/merged_data1.csv', index=False)
```

```python
df = pd.read_csv('Data/merged_data1.csv')
```

```python
# Define columns to check for all 0 or NaN values
columns_to_check = df.columns.difference(['ObjectId', 'Date'])

# Filter out rows where all entries in the columns_to_check are 0 or NaN
df_cleaned = df.loc[~df[columns_to_check].apply(lambda row: (row == 0) | (row.isna())).all(axis=1)]

```

```python
#change settings so all columns of the dataframe are displayed
pd.set_option('display.max_columns', None)
```
